# Common script to calculate the MHW for Arabian Sea, Bay of Bengal and the North Indian Ocean

```python
# common script to calculate the marine heat wave 
# import the packages and libraries
from glob import glob
from pathlib import Path
import numpy as np, pandas as pd, xarray as xr
import matplotlib.pyplot as plt
from dask.diagnostics import ProgressBar
import marineHeatWaves as mhw

# SST files path
sst_path = "/home/Desktop/Noah_data_1982-2024_SST_daily_mean/sst.day.mean.*.nc"
SST_VAR    = "sst"

# define the baseline year
BASELINE = (1982, 2024)

# Event definition based on Hobday et al. (2016)
MIN_DUR  = 5     # minimum event duration (days)
MAX_GAP  = 2     # join across gaps up to this many days

# Dask-friendly chunks
CHTIME, CHXY = 160, 80
 
# Set the regions
ROI_DICT = {
    "Arabian Sea": {"lon_min": 40.0, "lon_max": 80.0, "lat_min":  0.0, "lat_max": 30.0, "slug": "arabian_sea"},
    "Bay Of Bengal": {"lon_min": 80.0, "lon_max": 110.0,"lat_min":  0.0, "lat_max":  30.0, "slug": "bay_of_bengal"},
    "North Indian Ocean": {"lon_min": 40.0, "lon_max": 110.0,"lat_min":  0.0, "lat_max":  30.0,"slug": "north_indian_ocean"},
}

OUTROOT = Path("outputs_mhw_without_avg"); OUTROOT.mkdir(parents=True, exist_ok=True)

# open SST data using xarray
def open_sst(files_glob: str, roi: dict) -> tuple[xr.Dataset, str, str]:
    paths = sorted(glob(files_glob))
    if not paths:
        raise FileNotFoundError(f"No files match: {files_glob}")
    ds = xr.open_mfdataset(paths, combine="by_coords", chunks={"time": CHTIME}, engine="netcdf4")
    latn = "lat" if "lat" in ds.coords else "latitude"
    lonn = "lon" if "lon" in ds.coords else "longitude"
    ds = ds.sel({latn: slice(roi["lat_min"], roi["lat_max"]), lonn: slice(roi["lon_min"], roi["lon_max"])})
    return ds, latn, lonn


# use the hobdey's defination on region
def _detect_mask_1d(sst_1d: np.ndarray, thresh_1d: np.ndarray, min_dur: int = MIN_DUR, max_gap: int = MAX_GAP) -> np.ndarray:
    
    ok = np.isfinite(sst_1d) & np.isfinite(thresh_1d)
    exc = ok & (sst_1d >= thresh_1d)
    if not exc.any():
        return exc

    x = exc.astype(np.int8)
    n = x.size

    # join short gaps
    i = 0
    while i < n:
        if x[i] == 1:
            j = i + 1
            while j < n and x[j] == 1:
                j += 1
            g0 = j
            while g0 < n and x[g0] == 0:
                g0 += 1
            gap_len = g0 - j
            if gap_len > 0 and gap_len <= max_gap:
                x[j:g0] = 1
                j = g0
            i = j
        else:
            i += 1

    # enforce min duration
    y = x.copy()
    i = 0
    while i < n:
        if y[i] == 1:
            j = i + 1
            while j < n and y[j] == 1:
                j += 1
            if (j - i) < min_dur:
                y[i:j] = 0
            i = j
        else:
            i += 1

    return y.astype(bool)

def detect_mask_time(sst: xr.DataArray, thresh: xr.DataArray) -> xr.DataArray:
    
    return xr.apply_ufunc(_detect_mask_1d, sst, thresh, input_core_dims=[["time"], ["time"]], output_core_dims=[["time"]], vectorize=True, dask="parallelized", output_dtypes=[bool])

# per grid climatology & threshold 
def _clim_thresh_time_1d(ords_1d: np.ndarray, temp_1d: np.ndarray, y0: int, y1: int, pct: int, min_dur: int, max_gap: int) -> tuple[np.ndarray, np.ndarray]:
    
    # If mostly missing, return NaNs to avoid unstable fits
    if np.isfinite(temp_1d).sum() < 30:
        n = temp_1d.size
        return np.full(n, np.nan, float), np.full(n, np.nan, float)

    # MHW detection
    _, clim = mhw.detect(ords_1d.astype(int), temp_1d.astype(float),climatologyPeriod=[int(y0), int(y1)], pctile=int(pct), minDuration=int(min_dur),
                         joinAcrossGaps=True, maxGap=int(max_gap),)
    seas   = np.asarray(clim["seas"],   float)
    thresh = np.asarray(clim["thresh"], float)
    return seas, thresh

# compute time-aligned climatology and threshold for every grid
def build_grid_baseline(ds: xr.Dataset, latn: str, lonn: str,pctile: int = 90) -> tuple[xr.DataArray, xr.DataArray]:
    
    t_index = pd.to_datetime(ds["time"].values)
    y0 = max(BASELINE[0], t_index.year.min())
    y1 = min(BASELINE[1], t_index.year.max())
    ords_da = xr.DataArray(np.array([d.toordinal() for d in t_index], dtype=int), coords={"time": ds["time"]}, dims=["time"],).chunk({"time": -1})  # single time chunk for gufunc

    sst = ds[SST_VAR].chunk({"time": -1, latn: CHXY, lonn: CHXY})

    seas_t, thresh_t = xr.apply_ufunc(_clim_thresh_time_1d, ords_da, sst,input_core_dims=[["time"], ["time"]], output_core_dims=[["time"], ["time"]], vectorize=True, dask="parallelized",
                                      output_dtypes=[float, float], kwargs=dict(y0=int(y0), y1=int(y1), pct=int(pctile), min_dur=MIN_DUR, max_gap=MAX_GAP))
    seas_t   = seas_t.rename("seas_t")
    thresh_t = thresh_t.rename("thresh_t")
    return seas_t, thresh_t
```

**MHW detection per grid for regions: Change regions to calculate and detect the MHW in "Arabian Sea", "Bay of Bengal", "North Indian Ocean". The file savig path will follow with regions accordingly.**

```python
# This script detect and save the per-grid MHW and save in zarr files for Arabian Sea
REG = "Arabian Sea"
ROI = ROI_DICT[REG]

#  Open SST data
ds, latn, lonn = open_sst(FILES_GLOB, ROI)

#  Land mask & chunks
ocean = ds[SST_VAR].notnull().any("time")
sst   = ds[SST_VAR].where(ocean).chunk({"time": CHTIME, latn: CHXY, lonn: CHXY})

#  Per-grid climatology & 90th-threshold 
seas_t, thresh_t = build_grid_baseline(ds, latn, lonn, pctile=90)

# Detect Hobday events per grid 
# compute boolean mask for logic
evt_mask_bool = detect_mask_time(sst.chunk({"time": -1}),thresh_t.chunk({"time": -1})).rename("mhw_mask")  # bool

# NaN on land & keep float so NaNs persist to Zarr
evt_mask = evt_mask_bool.where(ocean).astype("float32")


# Daily metrics for future analysis
intensity = (sst - seas_t).rename("intensity")                
excess    = (sst - thresh_t).where(evt_mask == 1).rename("excess")  # only on event days

#  Yearly per-grid summaries
# compute starts on the boolean mask
starts_bool = (evt_mask_bool & ~(evt_mask_bool.shift(time=1, fill_value=False)))

# NaN on land, keep float dtype (so NaNs survive)
starts = starts_bool.where(ocean).astype("float32")

events_per_year_grid = (starts.groupby("time.year").sum("time").where(ocean).rename("events_per_year").astype("float32"))  
days_per_year_grid = (evt_mask.groupby("time.year").sum("time").where(ocean).rename("days_per_year").astype("float32"))
mean_intensity_year_grid = (intensity.where(evt_mask == 1).groupby("time.year").mean("time").rename("mean_intensity_year").astype("float32"))
max_intensity_year_grid = (intensity.where(evt_mask == 1).groupby("time.year").max("time").rename("max_intensity_year").astype("float32"))
cumulative_intensity_year_grid = (intensity.where(evt_mask == 1).groupby("time.year").sum("time").rename("cumulative_intensity_year").astype("float32"))

# SAVE (Zarr + CSV) 
out_dir = OUTROOT / ROI_DICT[REG]["slug"]; out_dir.mkdir(parents=True, exist_ok=True)

daily_ds = xr.Dataset({"mhw_mask": evt_mask, "intensity": intensity, "excess": excess,"seas_t": seas_t, "thresh_t": thresh_t},coords={"time": ds["time"], latn: ds[latn], lonn: ds[lonn]},
                       ).chunk({"time": CHTIME, latn: CHXY, lonn: CHXY})

# Keep only core coords to avoid Zarr alignment issues
keep = {"time", latn, lonn}
dropc = [c for c in daily_ds.coords if c not in keep]
if dropc:
    daily_ds = daily_ds.reset_coords(dropc, drop=True)

print("Writing Zarr outputs… (can take time)")
with ProgressBar():
    daily_ds.to_zarr(out_dir / "mhw_daily.zarr", mode="w", safe_chunks=False, align_chunks=True)
    events_per_year_grid.to_zarr(out_dir / "mhw_events_per_year_grid.zarr", mode="w", safe_chunks=False, align_chunks=True)
    days_per_year_grid.to_zarr(out_dir / "mhw_days_per_year_grid.zarr", mode="w", safe_chunks=False, align_chunks=True)
    mean_intensity_year_grid.to_zarr(out_dir / "mhw_mean_intensity_year_grid.zarr", mode="w", safe_chunks=False, align_chunks=True)
    max_intensity_year_grid.to_zarr(out_dir / "mhw_max_intensity_year_grid.zarr", mode="w", safe_chunks=False, align_chunks=True)
    cumulative_intensity_year_grid.to_zarr(out_dir / "mhw_cumulative_intensity_year_grid.zarr", mode="w", safe_chunks=False, align_chunks=True)

print("Arabian Sea — done.")
```

```python
# This script plots the MHW event and days bar graphs with trend lines per decades and per year using MHW zarr data of Arabian Sea
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm

# Define the time period
BASELINE = (1982, 2024)
REG = "Arabian Sea"

# Load the zarr files
ds_event_path = xr.open_zarr("/home//Jupyter files/arabian_sea/mhw_events_per_year_grid.zarr")
ds_days_path = xr.open_zarr("/home/Desktop/Jupyter files/arabian_sea/mhw_days_per_year_grid.zarr")

# Corrected coordinate extraction
latn = "lat" if "lat" in ds_event_path.coords else "latitude"
lonn = "lon" if "lon" in ds_event_path.coords else "longitude"

# Extract the specific variable array from the loaded Dataset
events_per_year = ds_event_path["events_per_year"]
days_per_year  =  ds_days_path["days_per_year"]



# Recreate the ocean mask (True for ocean, False for land)
ocean = events_per_year.notnull().any("year")

# X-axis year tick 
YEAR_TICK_STEP  = 3
YEAR_TICK_START = None
YEAR_TICK_END   = None
YEAR_TICK_ROT   = 0

# Area Weights Calculation
def area_weights_1d(ds_event_path: xr.Dataset, latn: str) -> xr.DataArray:
    return np.cos(np.deg2rad(ds_event_path[latn]))

def apply_year_ticks(ax, years, step=YEAR_TICK_STEP, start=YEAR_TICK_START, end=YEAR_TICK_END, rotate=YEAR_TICK_ROT):
    years = np.asarray(years, dtype=int)
    y_min, y_max = int(years.min()), int(years.max())
    a = y_min if start is None else int(start)
    b = y_max if end   is None else int(end)
    if (step is None) or (step == 0):
        ticks = np.unique(years)
    else:
        ticks = np.arange(a, b + 1, int(step), dtype=int)
    ax.set_xticks(ticks)
    ax.set_xticklabels([str(y) for y in ticks], rotation=rotate)

# calculte the Basin area avarage
# Calculate weights
w_lat = area_weights_1d(ds_event_path, latn)

# Area-weighted regional means (per year)
freq_region = (events_per_year.where(ocean)).weighted(w_lat).mean(dim=[latn, lonn]).compute()
days_region = (days_per_year.where(ocean)).weighted(w_lat).mean(dim=[latn, lonn]).compute()

total_events_region = float(freq_region.sum().values)
total_days_region   = float(days_region.sum().values)

print(f"Mean Total Events: {total_events_region:.1f}, days: {total_days_region:.1f} per grid cell in {REG}")

# Plot the graphs
years_freq = (freq_region.coords.get("year", None).values
              if "year" in freq_region.coords
              else freq_region.get_index("year").values).astype(int)
years_days = (days_region.coords.get("year", None).values
              if "year" in days_region.coords
              else days_region.get_index("year").values).astype(int)
```

**Note: To calculate the year trendline, just remove the 10 multiplication from the *plot_trendlines* fuctions in required trendline calculation and chamge the unit from decaede to year. Rest of the code stays same. Comment out the trend which one isn't necessary**
**This is valid for all the trendline plotting in the given code scripts**

```python
# Events and days per decade bar graphs with trends
def plot_trendlines(ax, x,y):
    y= np.asarray(y)
    # 1. OLS linear regression trend 
    slop_ols, intercept_ols, r_val_ols, p_ols, std_err_ols= stats.linregress(x,y)
    line_ols= (slop_ols*x) + intercept_ols
    # p value test
    p_ols_str = "<0.01" if p_ols< 0.01 else f"{p_ols:.3f}"
    label_ols =  f"Linear Regression Trend: +{slop_ols*10:.2f}/decade ($p={p_ols_str}$)"
    # plot the trend line 
    ax.plot(x, line_ols, color= "red", linestyle= '--', linewidth= 2, label= label_ols)
    
    # 2. Theil-Sen non-parametric trend
    ts_res= stats.mstats.theilslopes(y,x, 0.95)
    slop_ts= ts_res[0]
    intercept_ts= ts_res[1]
    line_ts = (slop_ts*x) + intercept_ts
    # mann-kendall significance test 
    tau, p_ts = stats.kendalltau(x,y)
    # p value test
    p_ts_str = "<0.01" if p_ts < 0.01 else f"{p_ts:.3f}"
    label_ts =  f"Theil-Sen Trend: +{slop_ts*10:.2f}/decade ($p={p_ts_str}$)"
    # plot the trend line 
    ax.plot(x, line_ts, color= "black", linestyle= '--', linewidth= 2, label= label_ts)
    
    # 3. Poisson generalized linear model (GLM) trend 
    # Statsmodels requires explicitly add a constant for the intercept
    X_glm = sm.add_constant(x)
    glm_model = sm.GLM(y, X_glm, family=sm.families.Poisson())
    glm_results = glm_model.fit()
    # p value test
    p_glm = glm_results.pvalues[1]
    p_glm_str = "< 0.01" if p_glm < 0.01 else f"{p_glm:.3f}"
    # plot the trend line
    line_glm = glm_results.predict(X_glm)
    
    # The GLM slop will be in change per decade.
    slope_coef = glm_results.params[1]
    mean_y = np.mean(y)
    decadal_increase = slope_coef * mean_y * 10
    label_glm = f"Poisson GLM: +{decadal_increase:.2f}/decade ($p={p_glm_str}$)"
    ax.plot(x, line_glm, color= "green", linestyle= '--', linewidth= 2, label= label_glm)

    # add legends
    ax.legend()

# plot the Events bar chart with all trend
fig, ax1 = plt.subplots(figsize=(12, 4))
ax1.bar(years_freq, freq_region.values, color="#5b8def", width=0.8)
# add trend line
plot_trendlines(ax1,years_freq,freq_region)
ax1.set_title(f" MHW Events Per Year in {REG} ({BASELINE[0]}–{BASELINE[1]})", fontweight= "bold")
ax1.set_xlabel("Year", fontweight= "bold")
ax1.set_ylabel("MHW Events Per Year", fontweight= "bold")
ax1.grid(True, ls="--", alpha=0.4)
ax1.text(0.5, 0.92, f"Total Events: {total_events_region:.1f}", transform=ax1.transAxes, ha="center")
apply_year_ticks(ax1, years_freq)  
plt.tight_layout() 
plt.show()

# plot the days bar chart with all trend
fig, ax2 = plt.subplots(figsize=(12, 4))
ax2.bar(years_days, days_region.values, color="#f6a300", width=0.8)
# add trend line
plot_trendlines(ax2,years_days, days_region)
ax2.set_title(f"MHW Days Per Year in {REG} ({BASELINE[0]}–{BASELINE[1]})", fontweight= "bold")
ax2.set_xlabel("Year", fontweight= "bold"); ax2.set_ylabel("MHW Days Per Year", fontweight= "bold")
ax2.grid(True, ls="--", alpha=0.4)
ax2.text(0.5, 0.92, f"Total MHW days: {total_days_region:.1f}", transform=ax2.transAxes, ha="center")
apply_year_ticks(ax2, years_days) 
plt.tight_layout()
plt.show()
```

**This script calculate the *Change Point Analysis (CPA)* of the MHW events, days and cumulative intensity in the region. Change the region and respective files to calcute CPA**

```python
# Change point analysis of Arabian Sea MHW events, days and cumulative intensity
import numpy as np
import matplotlib.pyplot as plt
import ruptures as rpt
import xarray as xr

REG = "Arabian Sea"

# Prepare the MHW event data for CPA
ds_event= xr.open_zarr("/home/Desktop/Jupyter files/arabian_sea/mhw_events_per_year_grid.zarr")
ds_days = xr.open_zarr("/home/Desktop/Jupyter files/arabian_sea/mhw_days_per_year_grid.zarr")
ds_cum_int= xr.open_zarr("/home/Desktop/Jupyter files/arabian_sea/mhw_cumulative_intensity_year_grid.zarr")

events_per_year_year= ds_event["events_per_year"]
days_per_year_year = ds_days["days_per_year"]
cum_intensity_year = ds_cum_int["cumulative_intensity_year"]
ocean = events_per_year_year.notnull().any("year")

# Corrected coordinate extraction
latn = "lat" if "lat" in events_per_year_year.coords else "latitude"
lonn = "lon" if "lon" in events_per_year_year.coords else "longitude"

# X-axis year tick
YEAR_TICK_STEP  = 3
YEAR_TICK_START = None
YEAR_TICK_END   = None
YEAR_TICK_ROT   = 0


def apply_year_ticks(ax, years, step=YEAR_TICK_STEP, start=YEAR_TICK_START, end=YEAR_TICK_END, rotate=YEAR_TICK_ROT):
    years = np.asarray(years, dtype=int)
    y_min, y_max = int(years.min()), int(years.max())
    a = y_min if start is None else int(start)
    b = y_max if end   is None else int(end)
    if (step is None) or (step == 0):
        ticks = np.unique(years)
    else:
        ticks = np.arange(a, b + 1, int(step), dtype=int)
    ax.set_xticks(ticks)
    ax.set_xticklabels([str(y) for y in ticks], rotation=rotate)


# Area Weights Calculation
def area_weights_1d(ds_event: xr.Dataset, latn: str) -> xr.DataArray:
    return np.cos(np.deg2rad(ds_event[latn]))

# calculte the Basin area avarage
# Calculate weights
w_lat = area_weights_1d(ds_event, latn)
freq_region = events_per_year_year.where(ocean).weighted(w_lat).mean(dim=[latn, lonn], skipna= True).compute()
days_region = days_per_year_year.where(ocean).weighted(w_lat).mean(dim=[latn, lonn], skipna= True).compute()
cum_intensity_region = cum_intensity_year.where(ocean).weighted(w_lat).mean(dim=[latn, lonn], skipna= True).compute()

# Extract 1D arrays for the PELT algorithm
years = freq_region.year.values
y_event = freq_region.values.astype(float)
y_days = days_region.values.astype(float)
y_cum_intensity = cum_intensity_region.values.astype(float)

# Execute PELT Algorithm (l2 model detects shifts in the Mean)
# MHW Events CPA
algo_event = rpt.Pelt(model="l2").fit(y_event)
pen_event = np.log(len(y_event)) * np.var(y_event) # BIC penalty
cpts_event_idx = algo_event.predict(pen=pen_event)[:-1]    
cpts_event_years = [years[i] for i in cpts_event_idx]

# MHW Days CPA
algo_days = rpt.Pelt(model="l2").fit(y_days)
pen_days = np.log(len(y_days)) * np.var(y_days) # BIC penalty
cpts_days_idx = algo_days.predict(pen=pen_days)[:-1]    
cpts_days_years = [years[i] for i in cpts_days_idx]

#Cumulative Intensity CPA
algo_cum_intensity = rpt.Pelt(model="l2").fit(y_cum_intensity)
pen_cum_intensity = np.log(len(y_cum_intensity)) * np.var(y_cum_intensity) # BIC penalty
cpts_cum_intensity_idx = algo_cum_intensity.predict(pen=pen_cum_intensity)[:-1]    
cpts_cum_intensity_years = [years[i] for i in cpts_cum_intensity_idx]


print(f"Detected Regime Shifts (MHW Events): {cpts_event_years}")
print(f"Detected Regime Shifts (MHW Days): {cpts_days_years}")
print(f"Detected Regime Shifts (MHW Cumulative Intensity): {cpts_cum_intensity_years}")

# plot segment means
def plot_segment_means(ax, x, y, breakpoints_idx, color, linestyle, line_label):
    start_idx = 0
    # Include the end of the array to plot the final segment
    all_breaks = list(breakpoints_idx) + [len(y)] 
    
    for i, end_idx in enumerate(all_breaks):
        segment_y = y[start_idx:end_idx]
        mean_val = np.mean(segment_y)
        
        # Determine the x-coordinates for the horizontal line
        x_start = x[start_idx]
        # Prevent out-of-bounds index for the very last point
        x_end = x[end_idx - 1] if end_idx <= len(x) else x[-1]
        
        # Plot the mean line
        label= f"{line_label} segment mean"
        ax.hlines(mean_val, xmin=x_start, xmax=x_end, colors=color, linestyles='solid', linewidth=2, linestyle="--", label= label)
        
        start_idx = end_idx


# Plot the Twin-Axis CPA Graph for MHW Events
fig, ax1 = plt.subplots(figsize=(12, 4))
# Plot MHW Events (Bar)
ax1.bar(years, y_event, color="#5b8def", alpha=0.8)
ax1.set_xlabel("Year", fontweight="bold")
ax1.set_ylabel("MHW Events Per Year", fontweight="bold")
apply_year_ticks(ax1, years)

# Overlay the Change Points
for cpt in cpts_event_years:
    ax1.axvline(x=cpt, color="black", linestyle="--", linewidth=2.5, label=f"MHW Shift ({cpt})")

# plot the segment means
plot_segment_means(ax1, years, y_event, cpts_event_idx, color= "green", linestyle="--", line_label= "MHW event")

# Combine Legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
by_label = dict(zip(labels_1, lines_1))
ax1.legend(by_label.values(), by_label.keys(), loc="upper left", framealpha=0.9)
ax1.set_title(f" Change points of MHW Events in {REG} ({years[0]}-{years[-1]})", fontweight="bold")
ax1.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()


# Plot the Twin-Axis CPA Graph for MHW days
fig, ax2 = plt.subplots(figsize=(12, 4))
# Plot MHW Days (Bar)
ax2.bar(years, y_days, color="#f6a300", alpha=0.8)
ax2.set_xlabel("Year", fontweight="bold")
ax2.set_ylabel("MHW Days Per Year", fontweight="bold")
apply_year_ticks(ax2, years)

# Overlay the Change Points
for cpt in cpts_days_years:
    ax2.axvline(x=cpt, color="black", linestyle="--", linewidth=2.5, label=f"MHW Shift ({cpt})")

# plot the segment means
plot_segment_means(ax2, years, y_days, cpts_days_idx, color= "green", linestyle="--", line_label= "MHW days")
# Combine Legends
lines_2, labels_2 = ax2.get_legend_handles_labels()
by_label = dict(zip(labels_2, lines_2))
ax2.legend(by_label.values(), by_label.keys(), loc="upper left", framealpha=0.9)
ax2.set_title(f" Change points of MHW Days in {REG} ({years[0]}-{years[-1]})", fontweight="bold")
ax2.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# Plot the Twin-Axis CPA Graph for MHW Cumulative Intensity
fig, ax3 = plt.subplots(figsize=(12, 4))
# Plot MHW Cumulative Intensity (Bar)
ax3.bar(years, y_cum_intensity, color="#009E73", alpha=0.8)
ax3.set_xlabel("Year", fontweight="bold")
ax3.set_ylabel("MHW Cumulative Intensity Per Year", fontweight="bold")
apply_year_ticks(ax3, years)

# Overlay the Change Points
for cpt in cpts_cum_intensity_years:
    ax3.axvline(x=cpt, color="black", linestyle="--", linewidth=2.5, label=f"MHW Shift ({cpt})")

# plot the segment means
plot_segment_means(ax3, years, y_cum_intensity, cpts_cum_intensity_idx, color="#5b8def", linestyle="--", line_label= "MHW cumulative intensity")

# Combine Legends
lines_3, labels_3 = ax3.get_legend_handles_labels()
by_label = dict(zip(labels_3, lines_3))
ax3.legend(by_label.values(), by_label.keys(), loc="upper left", framealpha=0.9)
ax3.set_title(f" Change points of MHW Cumulative Intensity in {REG} ({years[0]}-{years[-1]})", fontweight="bold")
ax3.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# This is the script to calculte the K-S test on the MHW Event, days and Cumulative intensity
import scipy.stats as stats
import numpy as np

#  Define the split, start and end years for MHW events
event_split_year = 2017
event_start_year = 1997
event_end_year = 2024
#  Define the split, start and end years for MHW days
days_split_year = 2022
days_start_year = 2012
days_end_year = 2024


years_freq = (freq_region.coords.get("year", None).values
              if "year" in freq_region.coords
              else freq_region.get_index("year").values).astype(int)
years_days = (days_region.coords.get("year", None).values
              if "year" in days_region.coords
              else days_region.get_index("year").values).astype(int)


# plit the data into two time periods for MHW events
period1_event = (years_freq >= event_start_year) & (years_freq <= event_split_year)
period2_event = (years_freq > event_split_year) & (years_freq <= event_end_year)

# plit the data into two time periods for MHW days
period1_days = (years_days >= days_start_year) & (years_days <= days_split_year)
period2_days = (years_days > days_split_year) & (years_days <= days_end_year)

# Extract the continuous data 
freq_region_period1 = freq_region.values[period1_event]
freq_region_period2 = freq_region.values[period2_event]
days_region_period1 = days_region.values[period1_days]
days_region_period2 = days_region.values[period2_days]

# Run the Two-Sample K-S Test

# for MHW Events
ks_statistic, p_value_ks = stats.ks_2samp(freq_region_period1, freq_region_period2)

# calculate the CDF cordinates for the two periods
def calculate_cdf(data):
    x= np.sort(data)
    y= np.arange(1, len(x)+1/ len(x))
    return x,y

# Calculate CDFs for both periods for MHW events
x1_event, y1_event = calculate_cdf(freq_region_period1)
x2_event, y2_event = calculate_cdf(freq_region_period2)

# # Calculate CDFs for both periods for MHW days
x1_days, y1_days = calculate_cdf(days_region_period1)
x2_days, y2_days = calculate_cdf(days_region_period2)

# find the maximum gap coordinate for plotting the K-S statistic MHW events
x_all_event= np.sort(np.concatenate([x1_event, x2_event]))
y1_all_event= np.searchsorted(np.sort(freq_region_period1), x_all_event, side='right') / len(freq_region_period1)
y2_all_event= np.searchsorted(np.sort(freq_region_period2), x_all_event, side='right') / len(freq_region_period2)
max_idx_event = np.argmax(np.abs(y1_all_event - y2_all_event))
x_d_max_event = x_all_event[max_idx_event]

# find the maximum gap coordinate for plotting the K-S statistic MHW days
x_all_days= np.sort(np.concatenate([x1_days, x2_days]))
y1_all_days= np.searchsorted(np.sort(days_region_period1), x_all_days, side='right') / len(days_region_period1)
y2_all_days= np.searchsorted(np.sort(days_region_period2), x_all_days, side='right') / len(days_region_period2)
max_idx_days = np.argmax(np.abs(y1_all_days - y2_all_days))
x_d_max_days = x_all_days[max_idx_days]

# Thisa block print the K-S resuts  for MHW events and days 
# MHW Events K-S test results
print("Two-Sample Kolmogorov-Smirnov Test Results for MHW Events:")
print(f"Period 1 ({event_start_year}-{event_split_year}) vs Period 2 ({event_split_year+1}-{event_end_year})")
print(f"K-S Statistic (D): {ks_statistic:.4f}")
print(f"P-value: {p_value_ks:.4e}")

if p_value_ks < 0.05:
    print(f"Conclusion: The distribution of MHW Events has shifted significantly between the {event_start_year}-{event_split_year} and {event_split_year+1}-{event_end_year} periods.")
else:
    print("Conclusion: No statistically significant shift in the overall distribution.")

# plotting the CDFs for MHW events
fig, ax1 = plt.subplots(figsize= (12,4))
ax1.step(x1_event, y1_event, where='post', label=f"Period 1: {event_start_year}-{event_split_year}", color="#009E73")
ax1.step(x2_event, y2_event, where='post', label=f"Period 2: {event_split_year+1}-{event_end_year}", color="#5b8def")
ax1.plot([x_d_max_event, x_d_max_event], [y1_all_event[max_idx_event], y2_all_event[max_idx_event]], color="red", linestyle="--", linewidth=2, 
          label=f"K-S Statistic gap:(D={ks_statistic:.4f}\np-value={p_value_ks:.4e})")
ax1.set_title(f"CDF of MHW Events Per Year in {REG} for Two Periods", fontweight="bold")
ax1.set_xlabel("MHW Events Per Year", fontweight="bold")
ax1.set_ylabel("Cumulative Probability (%)", fontweight="bold")
ax1.legend(loc="upper left", framealpha=0.9)
ax1.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# MHW Days K-S test results
ks_statistic, p_value_ks = stats.ks_2samp(days_region_period1, days_region_period2)

print("\nTwo-Sample Kolmogorov-Smirnov Test Results for MHW Days:")
print(f"Period 1 ({days_start_year}-{days_split_year}) vs Period 2 ({days_split_year+1}-{days_end_year})")
print(f"K-S Statistic (D): {ks_statistic:.4f}")
print(f"P-value: {p_value_ks:.4e}")

if p_value_ks < 0.05:
    print(f"Conclusion: The distribution of MHW Days has shifted significantly between the {days_start_year}-{days_split_year} and {days_split_year+1}-{days_end_year} periods.")
else:
    print("Conclusion: No statistically significant shift in the overall distribution.")


# plotting the CDFs for MHW days
fig, ax2 = plt.subplots(figsize= (12,4))
ax2.step(x1_days, y1_days, where='post', label=f"Period 1: {days_start_year}-{days_split_year}", color="#009E73")
ax2.step(x2_days, y2_days, where='post', label=f"Period 2: {days_split_year+1}-{days_end_year}", color="#5b8def")
ax2.plot([x_d_max_days, x_d_max_days], [y1_all_days[max_idx_days], y2_all_days[max_idx_days]], color="red", linestyle="--", linewidth=2, 
          label=f"K-S Statistic gap:(D={ks_statistic:.4f}\np-value={p_value_ks:.4e})")
ax2.set_title(f"CDF of MHW Days Per Year in {REG} for Two Periods", fontweight="bold")
ax2.set_xlabel("MHW Days Per Year", fontweight="bold")
ax2.set_ylabel("Cumulative Probability (%)", fontweight="bold")
ax2.legend(loc="upper right", framealpha=0.9)
ax2.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

**following scripts calcultes and plots the bar graphs MHW categories in the regions occupying the % area occupied in a year**

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# region and time  periods
REG = "Arabian Sea"
BASELINE = (1982, 2024)

# Load the DAILY Zarr data (We need daily data to calculate the exact multipliers)
daily_zarr_path = "/home/Desktop/Jupyter files/arabian_sea/mhw_daily.zarr"
ds_daily = xr.open_zarr(daily_zarr_path)

latn = "lat" if "lat" in ds_daily.coords else "latitude"
lonn = "lon" if "lon" in ds_daily.coords else "longitude"

# Recreate ocean mask
ocean = ds_daily["mhw_mask"].notnull().any("time")

# Calculate the Hobday Multiplier (M) 
print("Calculating Hobday Multipliers...")
# The threshold gap (Delta T)
thresh_gap = ds_daily["thresh_t"] - ds_daily["seas_t"]

# The Multiplier: Intensity / Threshold Gap
# We only care about days where an MHW is actively occurring (mhw_mask == 1)
multiplier = (ds_daily["intensity"] / thresh_gap).where(ds_daily["mhw_mask"] == 1)

# Find the MAXIMUM category reached by each grid cell for each year
annual_max_mult = multiplier.groupby("time.year").max("time").compute()

# --- 3. Categorize the Grid Cells ---
# Create boolean masks for each category based on Hobday et al. (2018)
cat1 = (annual_max_mult >= 1) & (annual_max_mult < 2) # Moderate
cat2 = (annual_max_mult >= 2) & (annual_max_mult < 3) # Strong
cat3 = (annual_max_mult >= 3) & (annual_max_mult < 4) # Severe
cat4 = (annual_max_mult >= 4)                         # Extreme

# Spatial Area Weighting
print("Calculating Spatial Area Fractions...")
def area_weights_1d(ds, latn):
    return np.cos(np.deg2rad(ds[latn]))

w_lat = area_weights_1d(ds_daily, latn)
ocean_weights = w_lat.where(ocean).broadcast_like(ocean)
total_ocean_area = ocean_weights.sum(dim=[latn, lonn])

# Calculate what percentage of the basin area falls into each category
frac_cat1 = (cat1 * ocean_weights).sum(dim=[latn, lonn]) / total_ocean_area * 100
frac_cat2 = (cat2 * ocean_weights).sum(dim=[latn, lonn]) / total_ocean_area * 100
frac_cat3 = (cat3 * ocean_weights).sum(dim=[latn, lonn]) / total_ocean_area * 100
frac_cat4 = (cat4 * ocean_weights).sum(dim=[latn, lonn]) / total_ocean_area * 100

years = frac_cat1.year.values

# Plotting the Stacked Bar Graph
fig, ax = plt.subplots(figsize=(14, 6))

# Standard MHW Category Colors (Hobday et al., 2018)
c1, c2, c3, c4 = "#ffc866", "#ff6900", "#9e0000", "#2d0000"

# Plot stacked bars
p1 = ax.bar(years, frac_cat1.values, color=c1, edgecolor="black", linewidth=0.3, label="Category I (Moderate)")
p2 = ax.bar(years, frac_cat2.values, bottom=frac_cat1.values, color=c2, edgecolor="black", linewidth=0.3, label="Category II (Strong)")
p3 = ax.bar(years, frac_cat3.values, bottom=(frac_cat1 + frac_cat2).values, color=c3, edgecolor="black", linewidth=0.3, label="Category III (Severe)")
p4 = ax.bar(years, frac_cat4.values, bottom=(frac_cat1 + frac_cat2 + frac_cat3).values, color=c4, edgecolor="black", linewidth=0.3, label="Category IV (Extreme)")

ax.set_title(f"Maximum MHW Category Spatial Footprint in {REG} ({BASELINE[0]}–{BASELINE[1]})", fontsize=14, fontweight="bold")
ax.set_xlabel("Year", fontsize=12, fontweight="bold")
ax.set_ylabel("Percentage of Basin Area (%)", fontsize=12, fontweight="bold")

# Y-axis is naturally bounded from 0 to 100%
ax.set_ylim(0, 100)

# Format Y-axis ticks
ax.set_yticks(np.arange(0, 101, 10))

# Format X-axis ticks
ticks = np.arange(int(years.min()), int(years.max()) + 1, 3)
ax.set_xticks(ticks)
ax.set_xticklabels([str(y) for y in ticks])

# Reverse legend order so Extreme is on top, matching the stack
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], loc='lower center', framealpha=0.9)
ax.grid(True, ls="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
